# UNIT 13 숫자 데이터를 쉽게 다루게 돕는 numpy 라이브러리  
matplotlib 홈페이지([matplotlib](https://matplotlib.org/stable/tutorials))는 데이터 시각화와 관련된 다양한 예시를 제공하고 있다. 아래는 그중 하나의 예시이다. numpy 라이브러리는 np라는 별명으로 부르는 것을 확인할 수 있다. 파이썬 리스트를 사용하여 같은 결과를 얻을 수 있지만  numpy를 사용한 것이 더 간결하다.  

```python
# using numpy library
import matplotlib.pyplot as plt
import numpy as np
t = np.arange(0., 5., 0.2)
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
# using python list
import matplotlib.pyplot as plt
t = []
p2 = []
p3 = []
for i in range(0, 50, 2):
    t.append(i / 10)
    p2.append((i / 10) ** 2)
    p3.append((i / 10) ** 3)
plt.plot(t, t, 'r--', t, p2, 'bs', t, p3, 'g^')
plt.show()
```

## numpy 라이브러리 시작하기  

```python
import numpy as np
print(np.sqrt(2))  # 제곱근(squreroot)
print(np.pi)  # 원주율(π)
print(np.sin(0))  # sin 함수
print(np.cos(np.pi))  # cos 함수
a = np.random.rand(5)  # numpy 라이브러리의 sub 라이브러리인 random 호출
print(a)
print(type(a))  # ndarray = N-Dimensional Array
print(np.random.choice(6, 10))  # 0~5 사이 수 중 무작위로 10개 선택(중복순열)
print(np.random.choice(10, 6, replace=False))  # 중복 없이 선택(순열)
print(np.random.choice(6, 10, p=[0.1, 0.2, 0.3, 0.2, 0.1, 0.1]))  # 확률 지정
```

## numpy array 생성하기  
numpy 라이브러리에는 ndarray라는 특별한 데이터 타입의 배열이 사용되는데, numpy를 사용하려면 이를 다루는 데 익숙해질 필요가 있다. ndarray는  파이썬 리스트와 비슷하지만 차이점이 있다.  
- 요소를 구분하는 쉼표가 없다.
- 한 가지 타입의 데이터만 저장할 수 있다.
- zeros(), ones(), eye() 등의 함수를 사용하여 배열을 초기화할 수 있다.

```python
import numpy as np
a = np.array([1, 2, 3, 4])
print(a[1], a[-1])  # a의 1번 인덱스 값, -1번 인덱스 값 출력
print(a[1:])  # a의 1번 인덱스를 기준으로 슬라이싱 결과 출력

print(np.array([1, 2, '3', 4]))  # 한 가지 타입의 데이터만 저장

print(np.zeros(10))  # 0으로 이루어진 크기가 10인 배열 생성
print(np.ones(10))  # 1로 이루어진 크기가 10인 배열 생성
print(np.eye(3))  # 3*3 크기의 단위 행렬 생성

print(np.arange(3))  # arange() 함수에 1개 값 입력: 0이상, 3 미만, 1씩 증가
print(np.arange(3, 7))  # arange() 함수에 1개 값 입력: 3이상, 7미만, 1씩 증가
print(np.arange(3, 7, 2))  # arange() 함수에 1개 값 입력: 3이상, 7미만, 2씩 증가

print(np.arange(1, 2, 0.1))  # 1이상, 2미만, .1씩 증가
print(np.linspace(1, 2, 11))  # 1이상, 2이하 구간을 11등분

print(np.zeros(10) + 5)  # 초깃값이 5인 배열 10개 생성
print(np.sqrt(np.linspace(1, 2, 11)))  # 배열 요소 하나씩 sqrt 적용
```

```python
# numpy 라이브러리를 사용한 sin함수 그래프 그리기
import matplotlib.pyplot as plt
import numpy as np
a = np.linspace(-np.pi, np.pi, 100)
plt.plot(a, np.sin(a))
plt.plot(a, np.cos(a))  # sin(a) x축 방향으로 π/2만큼 평행이동
plt.plot(a+(np.pi/2), np.sin(a))  # sin(a) x축 방향으로 -π/2만큼 평행이동 (= cos(a))
plt.show()
```

numpy array에는 조건에 부합하는 데이터만 선별적으로 저장하는 mask라는 기능이 있다.

```python
# mask 개요
import numpy as np
a = np.arange(-5, 5)
print(a)
print(a<0)  # 조건에 대한 Boolean data array 반환
print(a[a<0])  # True에 대응하는 인수들만 반환

mask1 = abs(a) > 3  # 변수에 저장 가능
mask2 = abs(a) % 2 == 0
print(a[mask1+mask2])  # mask1 OR mask2
print(a[mask1*mask2])  # mask1 AND mask2
```

```python
# numpy 라이브러리르 사용하여 버블차트 그리기
import matplotlib.pyplot as plt
import numpy as np
x = np.random.randint(-100, 100, 1000)
y = np.random.randint(-100, 100, 1000)
mask1 = abs(x) > 50
mask2 = abs(y) > 50
x = x[mask1+mask2]
y = y[mask1+mask2]
size = np.random.rand(len(x))*100
plt.scatter(x, y, s=size, c=x, cmap='jet', alpha=0.7)
plt.colorbar()
plt.show()
```
